```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_income = pd.read_csv('IncomeDataOntario.csv')

df_income
```

```python
df_income = df_income.drop(index=[15, 16])

df_income = df_income.rename(columns={"Income concept 6": "Income type"})
#df_income["Economic family type"] = df_income["Economic family type"].str.rstrip(" 7")
#df_income["Economic family type"] = df_income["Economic family type"].str.rstrip(" 8")
df_income["Economic family type"] = df_income["Economic family type"].str.strip(" 78")
df_income
```

```python
year_cols = ["2019", "2020", "2021", "2022", "2023"]

df_income[year_cols] = df_income[year_cols].replace({',':''}, regex=True)
df_income
```

```python
df_income = pd.melt(
    df_income, 
    id_vars=["Income type", "Economic family type"], 
    #value_vars=year_cols,
    var_name="Year",
    value_name="Income")

df_income
```

```python
#Median after-tax income is most closely represent renters income state when paying rent
#The median represents the most popular income amount, rather than mean
df_income = df_income[df_income["Income type"] == "Median after-tax income"]
#Persons not in an economic family represent majority of renters
#Rent prices are per person, so it most closely represents the renting audience
df_income = df_income[df_income["Economic family type"] == "Persons not in an economic family"]

base_income = float(df_income.loc[df_income["Year"] == "2019", "Income"].iloc[0])
df_income["Income Index"] = (df_income["Income"].astype(float) / base_income) * 100
df_income["Income Index"] = df_income["Income Index"].round(2)

df_income = df_income.drop(columns=["Income type", "Economic family type" ])

df_income
```

```python
df_rent_new = pd.read_csv('AllRentData.csv')

## Filter to only 2019 to 2025 data
df_rent_new = df_rent_new[(df_rent_new['REF_DATE'] >= 2019) & (df_rent_new['REF_DATE'] <= 2025)]

## Drop rows where VALUE is NaN
df_rent_new = df_rent_new.dropna(subset=["VALUE"])

df_rent_new = df_rent_new[df_rent_new['STATUS'] != 'F']

## Drop columns that are not needed
df_rent_new = df_rent_new.drop(columns=["DGUID", "UOM", "UOM_ID", "SCALAR_FACTOR", "SCALAR_ID", "VECTOR", "STATUS", "COORDINATE", "SYMBOL", "TERMINATED", "DECIMALS"])

## Filter to only include Ontario geography
df_rent_new = df_rent_new[df_rent_new['GEO'].str.contains("Ontario", case=False)]

## Drop duplicate rows
df_rent_new = df_rent_new.drop_duplicates()

## Rename columns
df_rent_new = df_rent_new.rename(columns={
    "REF_DATE": "Year",
    "GEO": "City",
    "VALUE": "Rent Price"
})

df_rent_new['City'] = df_rent_new['City'].str.split(',').str[0]

df_rent_new["Year"] = df_rent_new["Year"].astype(str)

df_rent_new = df_rent_new[df_rent_new["Rent Price"] != 0]

df_rent_new

df_rent_new.to_csv("NewRentData.csv", index=False)
```

```python
df_rent_new['Type of unit'].value_counts()
```

```python
# Value count of type of unit shows two bedroom units are the most popular
# Use two bedroom units to find median rent price to calculate growth index
df_rent_new = df_rent_new[df_rent_new["Type of unit"] == "Two bedroom units"]

df_rent_median = df_rent_new.groupby("Year")["Rent Price"].median().reset_index()

df_rent_median
```

```python
df_rent_median["Year"] = df_rent_median["Year"].astype(str)
df_income["Income"] = df_income["Income"].astype(int)

df_income = df_income.merge(df_rent_median, on="Year", how="left")

df_income["Year"] = df_income["Year"].astype(str)
base_rent = df_income.loc[df_income["Year"] == "2019", "Rent Price"].iloc[0]
df_income["Rent Index"] = (df_income["Rent Price"] / base_rent) * 100
df_income["Rent Index"] = df_income["Rent Index"].round(2)

df = df_income

df

df.to_csv("IncomeGrowth.csv", index=False)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["Year"], df["Income Index"], marker="o", color="orange", label='Income Index')
plt.plot(df["Year"], df["Rent Index"], marker="o", color="blue", label='Rent Index')
plt.xlabel("Year")
plt.ylabel("Growth Index")
plt.title("Income and Rent Growth (2019-2023)")
plt.grid(True)
plt.show()
```
